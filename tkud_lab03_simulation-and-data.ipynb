- Mức cao: chiều cao từ 175cm đến bé hơn 190cm 
- Mức rất cao: chiều cao từ 190cm trở lên


### Bài 9: Mô phỏng bốc ba lá bài

Một bộ bài tú lơ khơ có 52 lá. Với: 
    - Tập các nút là {2, 3,..10, J, Q, K, A}
    - Tập các chất là {'co', 'ro', 'chuong', 'bich`}

Bạn hãy mô phỏng:

a. Bốc 3 lá bài từ bộ bài trên. Lưu ý: không có trường hợp 1 bộ bài có hai lá trùng nhau

b. Tính điểm thu được (tổng điểm 3 lá)

c. Tìm hiểu cách hiển thị hình ảnh 3 lá bài đã bốc bằng python

````python
# YOUR CODE HERE

cards = [str(i) for i in range(2, 11)] + ['J', 'Q', 'K', 'A']
suits = ['heart', 'diamond', 'spade', 'club']
cards = [f'{card}-{suit}' for suit in suits for card in cards]

# a)
draw_3_cards = np.random.choice(cards, size=3, replace=False)
print(draw_3_cards)

# b)
# lá [2-10] tính điểm 2-10. [J, Q, K] tính điểm 10. [A] tính 11
point_map = {str(i): i for i in range(2, 11)}
point_map.update({'J': 10, 'Q': 10, 'K': 10, 'A': 11})
points = sum(point_map[card.split('-')[0]] for card in draw_3_cards)

print(f'Points: {points}')


# c)
import os
from PIL import Image

def display_cards(cards):
    images = []
    for card in cards:
        card_file = f"data/cards/{card}.jpg"
        if os.path.exists(card_file):
            img = Image.open(card_file)
            images.append(img)
        else:
            print(f"Không tìm thấy ảnh cho lá bài: {card}")

    plt.figure(figsize=(10, 4))
    for i, img in enumerate(images):
        plt.subplot(1, len(images), i + 1)
        plt.imshow(img)
        plt.axis("off")
    plt.show()


display_cards(draw_3_cards)
````


### Bài 10: Mô phỏng Monte Carlo để ước lượng π

Ước lượng giá trị của π bằng phương pháp Monte Carlo. Ý tưởng là tạo ngẫu nhiên các điểm trong một hình vuông và xác định tỷ lệ các điểm nằm trong một phần tư hình tròn.

Công việc:

Tạo 100.000 điểm ngẫu nhiên (x, y) trong một hình vuông đơn vị.
Đếm số điểm nằm trong phần tư hình tròn có bán kính 1.
Ước lượng π bằng công thức: 4 * (số điểm trong hình tròn) / (tổng số điểm).

````python
# YOUR CODE HERE
n_points = 100000

x = np.random.uniform(0, 1, n_points)
y = np.random.uniform(0, 1, n_points)

# Kiểm tra điểm nào nằm trong phần tư hình tròn (x^2 + y^2 <= 1)
inside_circle = (x**2 + y**2) <= 1

# Tính số điểm trong hình tròn
n_inside_circle = np.sum(inside_circle)

# Ước lượng π
estimated_pi = 4 * n_inside_circle / n_points

print(estimated_pi)
````
