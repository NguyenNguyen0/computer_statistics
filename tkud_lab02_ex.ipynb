## NUMPY

Bạn hãy phát sinh ngẫu nhiên mảng ndarray có tên `arr` gồm 100 phần tử số nguyên có giá trị từ 0..1000. Thực hiện các thao tác sau:

```python
import numpy as np
from numpy import random
arr = np.array(random.randint(1000, size=(100)))
```

### 1. Hiển thị các phần tử thỏa yêu cầu dưới đây

- 10 phần tử đầu tiên
- 5 phần tử cuối cùng
- Các phần tử thỏa mãn điều kiện chia hết cho 5 và lớn hơn 100

```python
print(arr[:10])

print(arr[-5:])

print(arr[(arr % 5 == 0) & (arr > 100)])
```

### 2. Cho biết mảng trên có bao nhiêu phần tử có giá trị nhỏ hơn 100. Hiển thị các phần tử đó

```python
smaller_100 = arr[arr < 100]

print(len(smaller_100))
print(smaller_100)
```

### 3. Thực hiện thống kê:

- Tính tổng mảng
- Tìm giá trị trung bình
- Tìm giá trị trung vị
- Tìm Q1, Q2, Q3
- Tìm phân vị 20 (percentile 20)

```python
print(np.sum(arr))

print(np.mean(arr))

print(np.median(arr))

print(np.quantile(arr, q=[0.25, 0.5, 0.75]))

print(np.percentile(arr, q=0.20))
```

### 4. Thực hiện việc chọn ngẫu nhiên (không thay thế) 10 phần tử từ mảng trên 5 lần. Hiển thị kết quả lựa chọn mỗi lần

```python
random.choice(arr, size=10, replace=False)
```

### 5. Tìm phần tử xuất hiện nhiều nhất trong mảng

```python
# unique + argmax
unique_elements, counts = np.unique(arr, return_counts=True)
print(unique_elements[np.argmax(counts)])

# stats from scipy
from scipy import stats
print(stats.mode(arr).mode)
```

### Câu 4: Định dạng đường thẳng

Phát sinh x là các điểm dữ liệu từ $[0, \pi]$. Bạn hãy vẽ đồ thị $y = sin(x)$ và $y = cos(x)$ như dưới đây.
Gợi ý: sử dụng hằng số pi và hàm sau:
- `np.pi` : số $\pi$
- `np.linspace()`
- `np.sin()`
- `np.cos()`

```python
# This cell contains output. Don't remove!
```

```python
# YOUR CODE HERE
# Nếu muốn như trong hình thì nhân đôi pi
# x = np.linspace(0, 2 * np.pi, 1000)
x = np.linspace(0, np.pi, 1000)

plt.plot(x, np.sin(x), label='sin(x)')
plt.plot(x, np.cos(x), label="cos(x)", ls="dashed")
plt.grid()
plt.title('Đồ thị hàm sin(x) và cos(x)')
plt.legend()
plt.show()
```

```python
# YOUR CODE HERE
values = random.normal(size=10000)

plt.figure(figsize=(7,6))
plt.suptitle("Đồ thị hàm sin(x) và cos(x)")

plt.subplot(2, 1, 1)
plt.hist(values, bins=200)
plt.ylabel("Mật độ")
plt.title('Histogram')

plt.subplot(2, 1, 2)
plt.boxplot(values, vert=False)
plt.title('Boxplot')

plt.tight_layout(h_pad=2)
plt.show()
```
